```python
# 1. 웹 크롤링 기본 Format 설명

# bs4에 있는 BeautifulSoup 모듈을 가져오겠다. BeautifualSoup은 Parsing을 위한 모듈
from bs4 import BeautifulSoup 

# client(파이썬)이 웹사이트에 요청하기 위한 모듈
import requests

# 원하는 페이지 1개의 링크를 입력한다.
url = 'https://search.shopping.naver.com/search/all?query=%EC%8A%A4%ED%8C%B8&bt=-1&frm=NVSCPRO'

# 그 url에 데이터를 요청하고 받아온다.
result = requests.get(url)

# 만약에 사이트에서 한글, 영어가 깨진다면 인코딩 문제이다
# result.encoding = 'utf8'

# 가져온 겺과값을 a에 넣어준다. html 정보
a = result.text

# 파이썬에서 이용가능하도록 해줌
soup = BeautifulSoup(a, 'html.parser')

# 실행
soup
```

```python
# 6. 네이버 쇼핑 100개 상품 데이터 보여주기(과정) - 불만족
from bs4 import BeautifulSoup
import requests

url = 'https://search.shopping.naver.com/search/all?query=%EC%8A%A4%ED%8C%B8&bt=-1&frm=NVSCPRO'
result = requests.get(url)
#result.encoding = 'utf8'

a = result.text
soup = BeautifulSoup(a, 'html.parser')
#어떤 데이터를 가져올까?
#바로 앞에 태그를 가져오고 값이 많은 경우 더 구체화를 한다.

name = soup.select('p.cont > a')

#CASE1) 해당 태그의 값 i → 태그에 대한 정보
for i in name :
    print(i)
```

```python
# 7. 네이버 쇼핑 100개 상품 데이터 보여주기(과정) - 불만족
from bs4 import BeautifulSoup
import requests

url = 'https://search.shopping.naver.com/search/all?query=%EC%8A%A4%ED%8C%B8&bt=-1&frm=NVSCPRO'
result = requests.get(url)
#result.encoding = 'utf8'

a = result.text
soup = BeautifulSoup(a, 'html.parser')
#어떤 데이터를 가져올까?
#바로 앞에 태그를 가져오고 값이 많은 경우 더 구체화를 한다.

name = soup.select('p.cont > a')

#CASE2) 해당 태그의 값 i.text → 제품명 길이가 길면 안보임
for i in name :
    print(i.text)
```

```python
# 8. 네이버 쇼핑 100개 상품 데이터 보여주기(과정) - 만족
from bs4 import BeautifulSoup
import requests

url = 'https://search.shopping.naver.com/search/all?query=%EC%8A%A4%ED%8C%B8&bt=-1&frm=NVSCPRO'
result = requests.get(url)
#result.encoding = 'utf8'

a = result.text
soup = BeautifulSoup(a, 'html.parser')
#어떤 데이터를 가져올까?
#바로 앞에 태그를 가져오고 값이 많은 경우 더 구체화를 한다.

name = soup.select('p.cont > a')

#CASE3) 해당 태그의 속성값 i['title'] → 정확한 값 나옴
for i in name :
    print(i['title'])
```

```python
# 9. 네이버 쇼핑 상위 5개 제품명 보여주기
from bs4 import BeautifulSoup
import requests

url = 'https://search.shopping.naver.com/search/all?query=%EC%8A%A4%ED%8C%B8&bt=-1&frm=NVSCPRO'
result = requests.get(url)
#result.encoding = 'utf8'

a = result.text
soup = BeautifulSoup(a, 'html.parser')
#어떤 데이터를 가져올까?
#바로 앞에 태그를 가져오고 값이 많은 경우 더 구체화를 한다.

name = soup.select('p.cont > a')

for i in range(5) :
    print(name[i]['title'])
```
